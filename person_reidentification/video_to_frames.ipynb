```python
import cv2
import os
from PIL import Image
```

```python
# Specify Video Path
video_path = r"C:\Users\bomma\Downloads\verilog_project_file\OneDrive\Desktop\CS_Fundamentals\summer_intern\person_reidentification\WhatsApp Video 2026-06-02 at 9.09.36 PM.mp4"

# Check if video file exists
if not os.path.exists(video_path):
    print(f"Error: Video file not found at {video_path}")
else:
    print(f"Video file found: {video_path}")
```

```python
# Create folder for extracted frames

frames_folder = "saved_frames"

os.makedirs(frames_folder, exist_ok=True)

print("Frames folder created successfully")
```

```python
# --------------------------------------------
# Open Video
# --------------------------------------------
cap = cv2.VideoCapture(video_path)

# Check video
if not cap.isOpened():
    print("Error Opening Video")
else:
    print("Video Opened Successfully")

# --------------------------------------------
# Video Information
# --------------------------------------------
fps = int(cap.get(cv2.CAP_PROP_FPS))

print("FPS:", fps)

frame_count = 0
saved_count = 0

# --------------------------------------------
# Save Frame Every 0.5 Seconds
# --------------------------------------------
time_interval = 0.5

frame_interval = int(fps * time_interval)

print("Frame Interval:", frame_interval)

# --------------------------------------------
# Read Video
# --------------------------------------------
while True:

    ret, frame = cap.read()

    if not ret:
        break

    frame_count += 1

    current_time = frame_count / fps

    # Save frames at interval
    if frame_count % frame_interval == 0:

        hrs = int(current_time // 3600)
        mins = int((current_time % 3600) // 60)
        secs = int(current_time % 60)

        timestamp = f"{hrs:02d}-{mins:02d}-{secs:02d}"

        saved_count += 1

        # Use frame counter to ensure unique filenames
        frame_name = f"{frames_folder}/frame_{saved_count:03d}_{timestamp}.jpg"

        cv2.imwrite(frame_name, frame)

        print(f"Saved Frame {saved_count}: {timestamp}")

# --------------------------------------------
# Release Video
# --------------------------------------------
cap.release()

print("\n========== EXTRACTION COMPLETED ==========")
print("Total Frames Read:", frame_count)
print("Frames Saved:", saved_count)
```

```python
# Display extracted frames using OpenCV

files_list = sorted(os.listdir(frames_folder))

print(f"Total extracted frames: {len(files_list)}\n")

# List ALL frames
print("All extracted frames:")
for i, file in enumerate(files_list, 1):
    print(f"{i}. {file}")

print("\n" + "="*50)
print(f"✓ Total frames saved: {len(files_list)}")
print("="*50)
```
